```python
import pandas as pd

df = pd.read_csv("final.csv")
```

```python
df.columns
```

```python
questions = df['Question'].astype(str).tolist()
answers = df['Answer'].astype(str).tolist()
```

```python
print(questions)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L12-v2')  # Lightweight and fast
question_embeddings = model.encode(questions, show_progress_bar=False)
```

```python
import faiss
import numpy as np

dimension = question_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(question_embeddings))
```

```python
def get_answer(user_question, top_k=1):
    user_embedding = model.encode([user_question])
    distances, indices = index.search(np.array(user_embedding), top_k)
    return [answers[idx] for idx in indices[0]]
```

```python
user_input = "What are the benefits of investing in real estate in India?"
print(get_answer(user_input)[0])
```
